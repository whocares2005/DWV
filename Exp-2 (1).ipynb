```python
from nltk.tokenize import RegexpTokenizer
from pdfminer.high_level import extract_text
from nltk.probability import FreqDist

#Extract the text from PDF file
text=extract_text('2010.00462.pdf')
#create an instance of tokenizer using NLTK RegexpTokenizer
tokenizer=RegexpTokenizer('\w+')
#Tokenize the text read from file
tokens=tokenizer.tokenize(text)
#Find frequency distribution
fd=FreqDist(tokens)
#Find worrds whose length is greater than 5 and frequency greater than 20
long_freq_words=[words for words in tokens if len(words)>5 and fd[words]>20]
```
